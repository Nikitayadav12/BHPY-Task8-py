```python
import datetime
import random

print("🤖 Welcome to ChatBot! Type 'bye' anytime to exit.\n")

user_name = input("Before we start, may I know your name? ")
print(f"Bot: Nice to meet you, {user_name}! 👋 Type 'help' if you need options.\n")


chat_history = []


jokes = [
    "Why don’t scientists trust atoms? Because they make up everything! 😆",
    "Did you know? Python was named after 'Monty Python', not the snake! 🐍",
    "Math fact: Zero is the only number that can’t be represented in Roman numerals."
]

while True:
    user_input = input(f"{user_name}: ").lower()
    chat_history.append(user_input)

    if any(greet in user_input for greet in ["hi", "hello", "hey"]):
        print(f"Bot: Hey {user_name}! How’s your day going?")
        
    elif "how are you" in user_input:
        print("Bot: I'm just code, but I'm functioning perfectly 😄 What about you?")
        
    elif "your name" in user_input:
        print("Bot: I'm RuleBot, your friendly rule-based chatbot 🤖")

    elif "help" in user_input:
        print("Bot: You can ask me about time, date, weather, math, or even a joke!")

    elif "time" in user_input:
        current_time = datetime.datetime.now().strftime("%I:%M %p")
        print(f"Bot: The current time is {current_time}")

    elif "date" in user_input or "day" in user_input:
        today = datetime.datetime.now().strftime("%A, %d %B %Y")
        print(f"Bot: Today is {today}")

    elif "weather" in user_input:
        print("Bot: I'm not connected to live weather data 🌦, but I hope it’s pleasant where you are!")

    elif "joke" in user_input or "fact" in user_input:
        print("Bot:", random.choice(jokes))

    elif "add" in user_input or "plus" in user_input:
        try:
            numbers = [int(s) for s in user_input.split() if s.isdigit()]
            print(f"Bot: The sum is {sum(numbers)}")
        except:
            print("Bot: Hmm, I couldn’t calculate that. Try like 'add 5 and 10'.")
    
    elif "bye" in user_input:
        print(f"Bot: Bye {user_name}! It was nice chatting with you. 👋")
        print("Bot: Here’s what we talked about today:")
        for i, msg in enumerate(chat_history, 1):
            print(f"  {i}. {msg}")
        break

    else:
        print("Bot: Hmm... I didn’t catch that. Type 'help' to see what I can do.")
```
